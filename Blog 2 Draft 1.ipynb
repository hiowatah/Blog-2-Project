```python
def masters_api_call(que,key):
    url = 'https://na1.api.riotgames.com/lol/league/v4/grandmasterleagues/by-queue/' + que + '?api_key=' + key
    response = requests.get(url)  
    return response
```

```python
def masters_information(api_call):
    for information in api_call:
        cursor.execute("""
        INSERT INTO masters_players
        (summonerid, summoner_name, wins, losses, leaguepoints, summoner_id)
        VALUES
        (?,?,?,?,?,?)
        """,(information['summonerId'], information['summonerName'], information['wins'],
             information['losses'], information['leaguePoints'], information['summonerId']))
```
